```python
# ================================================================================
# EXERCISE 1: STUDENT GRADE MANAGEMENT SYSTEM
# ================================================================================

# PROBLEM:
# Create a student grade tracking system that stores data and generates reports.

# REQUIREMENTS:

# 1. Data Structure:
#    Create a dictionary to store students with this structure:
#    {
#        "S001": {
#            "name": "Alice",
#            "grades": [85, 90, 78, 92],  # list of test scores
#            "attendance": 0.90,  # percentage as decimal (0-1)
#            "projects": {"proj1": 95, "proj2": 88}
#        }
#    }

# 2. Functions to Write:
#    - add_student(students, student_id, name)
#    - add_grade(students, student_id, grade)
#    - calculate_test_average(students, student_id)
#    - calculate_final_score(students, student_id)
#      Formula: Tests 60% + Projects 25% + Attendance 15%
#      (Attendance score = attendance * 100)
#    - get_letter_grade(score)
#      A: 90+, B: 80-89, C: 70-79, D: 60-69, F: <60
#    - save_to_file(students, filename)
#    - load_from_file(filename)

# 3. Create Data:
#    - Add at least 5 students with complete data
#    - Each student should have 4+ test grades
#    - Each student should have 2+ project scores

# 4. Generate Report:
#    Calculate and display:
#    - Each student's test average, project average, and final score
#    - Letter grade for each student
#    - Class average
#    - Top 3 performing students
#    - Students at risk (final score < 60)

# 5. Save Results:
#    - Save the report to a text file with formatted output

# EXPECTED OUTPUT FORMAT:
# === STUDENT GRADE REPORT ===

# Student: Alice (ID: S001)
#   Test Average: 86.25
#   Project Average: 91.50
#   Attendance: 90.0%
#   Final Score: 87.43
#   Letter Grade: B

# Class Statistics:
#   Class Average: 82.15
#   Top 3: Alice (87.43), Bob (85.20), Charlie (83.10)
#   At Risk: David (58.50)

# Report saved to: student_report.txt
```

```python
import json
Rawfile = "students_raw.txt"

def add_student(students, student_id, name):  
    if student_id not in students:
        students[student_id] ={}
        students[student_id]["name"] = name
    else:
         if "name" not in students[student_id] or students[student_id]["name"] == "":
            students[student_id]["name"] = name
 
def add_grade(students, student_id, grade):
    if student_id  in students and "grade" not in students[student_id] :
        students[student_id]["grades"] = grade
        students[student_id]["test_average"] = sum(grade) / len(grade)
 
def add_attendances(students, student_id, attendance) :    
    if student_id  in students and "attendance" not in students[student_id] :
        students[student_id]["attendance"] = attendance

def add_projects(students, student_id, projects) :    
    if student_id  in students and "projects" not in students[student_id] :
        students[student_id]["projects"] = projects
        students[student_id]["projects_avg"] =sum(projects.values()) / len(projects)
def get_grade(score):
    # A: 90+, B: 80-89, C: 70-79, D: 60-69, F: <60
    if score >= 90:
        return 'A'
    elif score >= 80 and score < 90:
        return 'B'
    elif score >= 70 and score < 80:
        return 'C'
    elif score >= 60 and score < 70:
        return 'D'
    else:
        return 'F'
def calculate_final_score(students, student_id):
      testAvg = students[student_id]["test_average"]
      projAvg = students[student_id]["projects_avg"] 
      attendance = students[student_id]["attendance"]
      
      students[student_id]["final_score"] = round((testAvg *0.60)+ (projAvg * 0.25) +(attendance * 15 ) ,2)
      students[student_id]["Letter_Grade"] = get_grade(students[student_id]["final_score"])
     
# def writerawdata(student):
#     with open(Rawfile, "a") as file:
#         file.write(str(student))
#         file.write('\n')

def writerawdata(student):
    with open(Rawfile, "a") as file:
        file.write(json.dumps(student))
        file.write('\n')

def get_studentDetails ():   
    student_id = input("Enter studentID : ")
    Name = input("Enter Student Name : ")
    Grade=[]
    for i in range(4):
        inval = int(input(f"Enter Grade{i+1} : "))
        Grade.append(inval)
    Attendance = round(float(input("Enter attendance (between 0 and 1): ")), 2)
    project = {}
    for i in range(2):
        projname= "proj"+ str(i + 1)
        projvalue = int(input(f"Enter {projname} Score :"))
        project[projname] = projvalue

    #function call 
    add_student(students, student_id, Name)
    add_grade(students, student_id, Grade)
    add_attendances(students, student_id, Attendance)
    add_projects(students, student_id, project)
    calculate_final_score(students, student_id)
  
for i in range(2):
    students={}
    get_studentDetails ()
    print(students) 
    writerawdata(students)
```

```python
# ================================================================================
# EXERCISE 2: INVENTORY MANAGEMENT SYSTEM
# ================================================================================

# PROBLEM:
# Build an inventory tracking system that monitors stock levels and predicts 
# when to reorder products.

# REQUIREMENTS:

# 1. Data Structure:
#    Create inventory as a dictionary:
#    {
#        "PROD001": {
#            "name": "Laptop",
#            "category": "Electronics",
#            "price": 999.99,
#            "stock": 50,
#            "daily_sales": [5, 3, 7, 4, 6]  # last 5-7 days of sales
#        }
#    }

# 2. Functions to Write:
#    - add_product(inventory, product_id, name, category, price, stock)
#    - record_sale(inventory, product_id, quantity)
#      Updates stock and adds to daily_sales list
#    - calculate_sales_velocity(inventory, product_id)
#      Return average daily sales using the daily_sales list
#    - predict_stockout_days(inventory, product_id)
#      Calculate days until out of stock: stock / sales_velocity
#    - get_low_stock_products(inventory, threshold_days=7)
#      Return list of products that will run out within threshold days
#    - calculate_category_sales(inventory)
#      Return total units sold per category
#    - generate_reorder_report(inventory)

# 3. Create Data:
#    - Add at least 8 products across 3 categories (Electronics, Accessories, Office)
#    - Each product should have 5-7 days of sales history
#    - Vary stock levels (some low, some high)

# 5. Generate Report:
#    Display:
#    - Products needing reorder (< 7 days of stock remaining)
#    - Sales velocity for each product
#    - Total sales by category
#    - Products with highest/lowest sales velocity

# 6. Save Results:
#    - Save current inventory state to CSV

# EXPECTED OUTPUT FORMAT:
# === INVENTORY REORDER REPORT ===

# Products Needing Reorder (< 7 days stock):
#   PROD001: Laptop
#     Current Stock: 12 units
#     Sales Velocity: 4.8 units/day
#     Days Until Stockout: 2.5 days
#     Recommended Reorder: 50 units

#   PROD004: Mouse
#     Current Stock: 18 units
#     Sales Velocity: 2.9 units/day
#     Days Until Stockout: 6.2 days
#     Recommended Reorder: 30 units

# Category Performance:
#   Electronics: 142 units sold
#   Accessories: 87 units sold
#   Office: 64 units sold

# Top Selling Products:
#   1. Laptop (velocity: 4.8/day)
#   2. USB Cable (velocity: 3.2/day)
#   3. Mouse (velocity: 2.9/day)



def add_product(inventory, product_id, name, category, price, stock): 
    inventory[product_id] = {}
    inventory[product_id]["name"]=name
    inventory[product_id]["category"]=category
    inventory[product_id]["price"]=price
    inventory[product_id]["stock"]=stock

def record_sale(inventory, product_id, quantity):
    inventory[product_id]["daily_sales"]=quantity

def calculate_sales_velocity(inventory, product_id):
     daliySale = inventory[product_id]["daily_sales"]
     DaliySaleAvg = round(float(sum(daliySale)/len(daliySale)),2)
     return DaliySaleAvg

def predict_stockout_days(inventory, product_id):
     stockout_days = inventory[product_id]["stock"] / inventory[product_id]["salesvelocity"]
     return stockout_days

def getdetails():
    productID  = input("Enter ProductID : ")
    ProductName  = input("Enter Product Name : ")
    ProductCategory = input("Enter Product Category : ")
    ProductPrice = round(float(input("Enter Product Price : ")),2)
    ProductStock = int(input("Enter Product Stock : "))
    salerecords=[]
    for i in range(5):
        saleCount = int(input(f"Enter Day{i} sale "))
        salerecords.append(saleCount)

    #function call
    add_product(inventory,productID,ProductName,ProductCategory,ProductPrice, ProductStock)
    record_sale(inventory, productID, salerecords)
    salesvelocity= calculate_sales_velocity(inventory, productID)
    inventory[productID]["salesvelocity"] = salesvelocity
    inventory[productID]["stockoutdays"] = int( predict_stockout_days(inventory, productID) )


inventory = {}
for i in range(2):  
    getdetails()

print(inventory)
 
def get_low_stock_products(inventory, threshold_days=7):
     LowStockProducts = []
     for ProdID , i in inventory.items():
        if i['stockoutdays'] <threshold_days:
             LowStockProducts.append(i['name'])
     return LowStockProducts



def calculate_category_sales(inventory):
    category_sales = {}

    for pid, product in inventory.items():
         total_sold = sum(product["daily_sales"])
         category = product["category"]

         if category not in category_sales:
             category_sales[category] = 0
         category_sales[category] += total_sold

    return category_sales



def generateReport(inventory):
     lowStocks= get_low_stock_products(inventory, threshold_days=7)
     print(lowStocks)

     for pid, product in inventory.items():
        print(f" {pid} - {product['name']}: {product['salesvelocity']} units/day")

     category_sales = calculate_category_sales(inventory)
     for category, total in category_sales.items():
         print(f" {category}: {total} units sold")

     highest = max(inventory.items(), key=lambda x: x[1]["salesvelocity"])
     lowest = min(inventory.items(), key=lambda x: x[1]["salesvelocity"])
     print(f" {highest[0]} - {highest[1]['name']} ({highest[1]['salesvelocity']} units/day)")
     print(f" {lowest[0]} - {lowest[1]['name']} ({lowest[1]['salesvelocity']} units/day)")

import csv
 
def save_inventory_csv(inventory, filename="inventory_output.csv"):
    with open(filename, "w", newline="") as file:
        writer = csv.writer(file)
 
        writer.writerow(["Product ID", "Name", "Category", "Price",
                "Stock", "Daily Sales", "Sales Velocity", "Stockout Days"])
        for pid, product in inventory.items():
             writer.writerow([
                                pid,
                                product["name"],
                                product["category"],
                                product["price"],
                                product["stock"],
                                product["daily_sales"],
                                product["salesvelocity"],
                                product["stockoutdays"]
                            ])
 
    print(f"\nInventory saved to {filename}")
 
save_inventory_csv(inventory)
generateReport(inventory)
```
