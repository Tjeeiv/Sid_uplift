```python
# Exercise 1: Create a simple tip calculator
# Ask for bill amount and tip percentage
# Display the tip amount and total formatted to 2 decimal places

# Your code here:

billAmt = float(input("Enter the bill amount : "))
tipPer = float(input("Enter the Tip Percentage : "))
tipAmt = billAmt*(tipPer/100)
TotalBill = billAmt + tipAmt

print(f"Tip Amount: ₹ {tipAmt:.2f}")
print(f"Total Amount: ₹ {TotalBill:.2f}")
####################################################################
```
