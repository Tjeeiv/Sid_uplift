```python
# Exercise 4:

# Write code that:
# 1. Opens a file "temperatures.txt" which contains one temperature per line.
# 2. Reads all lines using readlines().
# 3. Converts them to floats.
# 4. Uses a function `average_temp(data)` to compute the average.
# 5. Prints:
# - The average temperature
# - "High" if avg > 25
# - "Moderate" if 15 <= avg <= 25
# - "Low" if avg < 15

# Your code here:
def average_temp(data):
    return sum(data) / len(data)

with open("temperatures.txt", "r") as file:
    lines = file.readlines()

temps = [float(line.strip()) for line in lines]

avg = average_temp(temps)
print("Average Temperature:", avg)

if avg > 25:
    print("High")
elif 15 <= avg <= 25:
    print("Moderate")
else:
    print("Low")
```

```python
# Exercise 5:

# Write a program that:
# 1. Opens a file "logs.txt" and reads each line.
# 2. Each line is formatted like: "LEVEL: message"
# 3. Use a function `count_level(filename, level)` that:
# - Opens the file
# - Counts how many lines start with the given level (e.g., "ERROR")
# - Uses loops + if to check matching lines
# 4. Print how many ERROR, WARNING, and INFO lines are in the file.

# Your code here: 

def count_level(filename, level):
    count = 0
    with open(filename, "r") as file:
        for line in file:
            if line.startswith(level + ":"):
                count += 1
    return count

 
filename = "logs.txt"
 
error_count = count_level(filename, "ERROR")
warning_count = count_level(filename, "WARNING")
info_count = count_level(filename, "INFO")
 
print("ERROR count:", error_count)
print("WARNING count:", warning_count)
print("INFO count:", info_count)
```
