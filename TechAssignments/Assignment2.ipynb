```python
# Exercise 2: 
# Create a dictionary for one department with keys "budget" and "employees"
# "employees" should be a list of two dictionaries, each with "name" and "salary"
# Example:
#     {
#        "Engineering": {
#            "budget": 500000,
#            "employees": [
#                {"name": "Alice", "salary": 90000},
#                {"name": "Bob", "salary": 85000}
#            ]
#        },
#        "Marketing": { ... }
#    }
# Calculate and print the total salary, then print if it is within the budget

# Your code here:

department ={
       "Engineering": {
           "budget": 500000,
           "employees": [
               {"name": "Alice", "salary": 90000},
               {"name": "Bob", "salary": 85000}
           ]
       },
       "Marketing": {
           "budget": 500000,
           "employees": [
               {"name": "Jon", "salary": 300000},
               {"name": "Snow", "salary": 300000}
           ] }
   }


for dept_name, dept_data in department.items():
     
    salaries = [emp["salary"] for emp in dept_data["employees"]]
    total_salary = sum(salaries)
    print(dept_name)
    if dept_data["budget"] > total_salary:
        print("Salary is within the budget")
    else:
        print("Salary is not within the budget")


####################################################################
```

```python
# Exercise 4:
# Create two sets: tech_subscribers and news_subscribers with some overlapping emails
# Print the emails that are in both sets
# Print all unique emails from both sets
# Remove one email from both sets and print the updated sets

# Your code here:

tech_subscribers = {"aaa@g.com","bbb@g.com", "xxx@g.com"}
news_subscribers = {"zzz@g.com","yyy@g.com", "xxx@g.com"}

print(f"Emails that are in both sets : {tech_subscribers & news_subscribers}")
print(f"unique set : {tech_subscribers | news_subscribers}")

removemail = (tech_subscribers & news_subscribers).pop()

tech_subscribers.remove(removemail)
news_subscribers.remove(removemail)

print(tech_subscribers)
print(news_subscribers)
```
